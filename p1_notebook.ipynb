```python
import pandas as pd
```

```python
# Import cell-count csv
cell_counts = pd.read_csv("cell-count.csv", index_col=False)
```

```python
cell_counts.head()
```

```python
cell_counts['total'] = cell_counts['b_cell'] + cell_counts['cd8_t_cell'] + cell_counts['cd4_t_cell'] + cell_counts['nk_cell'] + cell_counts['monocyte']
```

```python
cell_counts = cell_counts.loc[:, ['sample', 'b_cell', 'cd8_t_cell', 'cd4_t_cell', 'nk_cell', 'monocyte', 'total']]
```

```python
cell_counts.head()
```

```python
cell_counts = cell_counts.melt(id_vars=['sample', 'total'], var_name='population', value_name='count')
```

```python
cell_counts['count']
```

```python
cell_counts['percentage'] = 100 * (cell_counts['count'] / cell_counts['total'])
```

```python
cell_counts.head()
```

```python
cell_counts.to_csv("cell_count_transformed.csv", index=False)
```
